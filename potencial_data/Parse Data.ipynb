```python
def get_eolics(file):
    df  = pd.read_excel(file,header=1,sheetname="gen_real")
    df.index = list(range(len(df)))
    date = df.columns[0]
    if type(date) is not datetime.datetime:
        df  = pd.read_excel(file,header=2,sheetname="gen_real")
        df.index = list(range(len(df)))
        date = df.columns[0]
    df.columns = list(range(len(df.columns)))
    df = df[list(range(1,len(df.columns)))]
    df.columns=df.iloc[0].values
    df = df.iloc[1:]
    
    cut = np.where((df["Centrales"] == "Total Generación SIC").values)[0][0]
    solares_cut = np.where((df["Centrales"] == "Solares").values)[0]
    eolicas_cut =  np.where((df["Centrales"] == "Eólicas").values)[0][0]
    if len(solares_cut) == 0:
        data_eolicas = df[eolicas_cut+1:cut]
    else:
        data_eolicas = df[eolicas_cut+1:solares_cut[0]]
        
    data_eolicas.index = list(range(len(data_eolicas)))
    return date,data_eolicas
```

```python
for year in range(2007,2017):
    files = glob.glob("excel/{}/OP*.xls".format(year))
    os.makedirs("eolic_csv/{}".format(year),exist_ok=True)
    for file in files:
        try:
            date,df = get_eolics(file)
            out_file = "eolic_csv/{}/{}.csv".format(year,date.strftime("%Y-%m-%d"))
            print("Correct: ",out_file)
            df.to_csv(out_file)
        except:
            print("Error",file)
print("Finish")
```
