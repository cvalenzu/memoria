```python
b08_viento = b08["viento"].as_matrix()

def getDataWindowed(data,inSize):
    matrix = np.zeros((len(data)-inSize, inSize))
    for i in range(len(data)-inSize):
        matrix[i,:] = data[i+1:i+inSize+1]
    return matrix

data = getDataWindowed(b08_viento,12)
```

```python
figure(10).clear()
plot(data[0:100,0])
title('A sample of data')
```

```python
# load the data
dataLen = len(data)
trainLen = int(dataLen*0.7)
testLen = int(dataLen*0.3)
initLen = 100
```

```python
# generate the ESN reservoir
inSize = outSize = 12
resSize = 1000
a = 0.3 # leaking rate

random.seed(42)
Win = (random.rand(resSize,1+inSize)-0.5) * 1
W = random.rand(resSize,resSize)-0.5
# Option 1 - direct scaling (quick&dirty, reservoir-specific):
#W *= 0.135 
# Option 2 - normalizing and setting spectral radius (correct, slow):
print 'Computing spectral radius...',
rhoW = max(abs(linalg.eig(W)[0]))
print 'done.'
W *= 1.25 / rhoW

# allocated memory for the design (collected states) matrix
X = zeros((1+inSize+resSize,trainLen-initLen))

# set the corresponding target matrix directly
Yt = data[None,initLen+1:trainLen+1]

# run the reservoir with the data and collect X
x = zeros(resSize)
for t in range(trainLen-outSize):
    u = data[t:t+outSize]
    x = (1-a)*x + a*tanh( dot( Win, hstack((1,u)) ) + dot( W, x ) )
    if t >= initLen:
        X[:,t-initLen] = hstack((1,u,x))
```

```python
# train the output
reg = 1e-8  # regularization coefficient
X_T = X.T
#Wout = dot( dot(Yt,X_T), linalg.inv( dot(X,X_T) + reg*eye(1+inSize+resSize) ) )
Wout = dot( Yt, linalg.pinv(X) )
```

```python
# run the trained ESN in a generative mode. no need to initialize here, 
# because x is initialized with training data and we continue from there.
Y = zeros((outSize,testLen))
u = data[trainLen:trainLen+outSize]
for t in range(testLen-outSize):
    x = (1-a)*x + a*tanh( dot( Win, hstack((1,u)) ) + dot( W, x ) )
    y = dot( Wout, hstack((1,u,x)) )
    Y[:,t:t+outSize] = y
    print y
    # generative mode:
    #u = y
    ## this would be a predictive mode:
    u = data[trainLen+t+1:trainLen+t+outSize+1] 
```

```python

# compute MSE for the first errorLen time steps
errorLen = 500
mse = sum( square( data[trainLen+1:trainLen+errorLen+1] - Y[0,0:errorLen] ) ) / errorLen
print 'MSE = ' + str( mse )

# plot some signals
figure(1).clear()
plot( data[trainLen+1:trainLen+testLen+1], 'g' )
plot( Y.T, 'b' )
title('Target and generated signals $y(n)$ starting at $n=0$')

legend(['Target signal', 'Free-running predicted signal'])

figure(2).clear()
plot( X[0:20,0:200].T )
title('Some reservoir activations $\mathbf{x}(n)$')

figure(3).clear()
bar( range(1+inSize+resSize), Wout.T )
title('Output weights $\mathbf{W}^{out}$')

show()
```
